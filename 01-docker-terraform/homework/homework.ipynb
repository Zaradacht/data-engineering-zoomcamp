```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
query = """
SELECT * FROM green_tripdata
"""

df = pd.read_sql(query, con=engine)
df
```

```python
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'])
df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'])
df['lpep_pickup_datetime'] = df['lpep_pickup_datetime'].dt.date
df['lpep_dropoff_datetime'] = df['lpep_dropoff_datetime'].dt.date
df
```

```python
from datetime import datetime
date_filter = datetime.strptime('2019-09-18', '%Y-%m-%d').date()
filtered = df[(df['lpep_pickup_datetime'] == date_filter) & (df['lpep_dropoff_datetime'] == date_filter)]
len(filtered)
```

```python
query = """
SELECT 
	index,
	lpep_pickup_datetime,
	MAX(trip_distance) OVER (ORDER BY trip_distance DESC) AS max
FROM green_tripdata
LIMIT 1 
"""

df = pd.read_sql(query, con=engine)
df
```
